Recordemos que, antes de estos pasos, debíamos separar los datos haciendo cross-validation, es decir, segmentarlos en training y testing, en este caso

```python
from sklearn.model_selection import train_test_split
```

- Dividimos en train y test
- Convertimos los 4 segmentos de data y labels a objetos `pytorch` con TensorDataset
- Por último, agrupamos los datos en lotes (batches) aleatoriamente esparcidos usando DataLoader

## Separando los datos y tranformándolos a objetos de `PyTorch`

Recordemos el procedimiento:

- Dividimos los datos usando cross validation

- Los convertimos a objetos de pytorch usando `TensorDataset` (transformando previamente las observaciones y etiquetas a tensores si no lo estaban)

- Utilizamos dataloaders para agrupar en lotes (batches) los datos aleatoriamente esparcidos, y en batch_size=1 para el conjunto de testing

```python
#separar datos
train_data,test_data, train_labels,test_labels = train_test_split(data, labels, test_size=0.2)

#convertir a pytorch datasets 
#recuerda que incluye las etiquetas
train_data = TensorDataset(train_data, train_labels)
test_data = TensorDataset(test_data, test_labels)

#convertirlos a objetos DataLoader
#usaremos batch_size=12 para los datos de entrenamiento
train_loader = DataLoader(train_data, shuffle=True, batch_size=12)
#el conjunto de testing siempre tendrá un tamaño de lote de 1 ya que las predicciones son individuales (único batch)
test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0]) #igual al número total de observaciones de testing
```

```python
for batch_samples, batch_labels in train_loader:
    print(batch_samples.shape, batch_labels.shape)

batch_samples, batch_labels #un ejemplo de la última iteración
```

```python
for sample, label in test_loader:
    print(sample.shape, label.shape)
```

## Creando el modelo

```python
class Model(nn.Module):
    
    def __init__(self):
        super().__init__()
        
        self.inp = nn.Linear(4,64)
        self.hid = nn.Linear(64,64)
        self.out = nn.Linear(64, len((test_labels).unique()))
                             
    def forward(self, x):
        
        x = F.relu(self.inp(x))
        x = F.relu(self.hid(x))
        x = self.out(x)
        
        return x                       
```

```python
def build_model():
    
    model = Model()
    loss_func = nn.CrossEntropyLoss() #ver docs (usa LogSoftmax debido a la propiedad de los logaritmos de hacer los números pequeños "más espaciados")
    #por lo que añade el logaritmo a la función softmax

    learning_rate = 0.01

    optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
    
    return model, loss_func, optim
```

## Entrenando el modelo (entramiento en mini-batches)

Pasamos cada lote de observaciones de entrenamiento, realizamos backprop, hasta haber pasado todos los lotes de esta manera. Una vez hecho esto, se habrá completado 1 epoch correctamente

```python
n_epochs = 500


def train():
    
    #no almacenaremos las pérdidas, tan solo las precisiones
    train_accs = []
    test_accs = []
    
    for epoch in range(n_epochs): #tqdm
        
        #para cada epoch, alimentaremos los datos al modelo en lotes de 12 observaciones
        batches_accs = [] #para cada lote, almacenaremos las precisión del modelo
        
        for X_batch, y_batch in train_loader: #actualizamos los pesos para cada lote, hasta pasar por todos los lotes = 1 epoch
            
            #forward pass
            y_hat = model(X_batch)

            #compute error
            loss = loss_func(y_hat, y_batch)
            
            #backward pass
            optim.zero_grad() #clear previous gradients cache
            loss.backward() #compute gradients based on the error
            optim.step() #update parameters

            #calculamos la precisión
            batch_acc = 100*torch.mean(((torch.argmax(y_hat, axis=1) == y_batch).float())).item()
            
            batches_accs.append(batch_acc)
            #esta implementación de mini-batch ayuda a una mayor rapidez de cómputo y mejor generalización del modelo
        
        #una vez esten almacenadas todas las precisiones para cada batch de la epoch actual
        #almacenaremos la media de las precisiones en dicha epoch
        train_accs.append(np.mean(batches_accs))
        
        
        #calculamos la precisión en el conjunto de testing una vez completada la epoch
        X_batch_test, y_batch_test = next(iter(test_loader)) #accedemos a los datos de testing del DataLoader
        #recuerda que al ser un único batch para testing, usamos next(iter()) en el generador
        #pero si fuera el conjunto de entrenamiento, esas funciones tan solo nos darían el primer elemento (mini-batch)
        
        preds = torch.argmax(model(X_batch_test), axis=1)
        
        test_accs.append(100*torch.mean((preds==y_batch_test).float()).item())
        
    return train_accs, test_accs
```

## Entrenando el modelo

```python
model, loss_func, optim = build_model()

train_accs, test_accs = train()
```

```python
plt.figure(figsize=(10,6))
plt.plot(train_accs, "ro-");
plt.plot(test_accs, "ys-");
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)");
plt.legend(["Train", "Test"]);
```
