# Running Parametric Experiments on GD (1D)

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
plt.style.use("dark_background")
```

```python
x = sp.symbols("x")
```

```python
expr = sp.sin(x) * sp.exp(-x**2*.05)
```

```python
expr
```

```python
func = sp.lambdify(x, expr, "numpy") 
```

```python
deriv = sp.diff(expr)
deriv_func = sp.lambdify(x, deriv, "numpy")
```

```python
deriv
```

### Definimos el vector de 401 valores linealmente espaciados desde -2*$\pi$ hasta 2*$\pi$.

### A continuación, visualizamos la función

$$\large f(x) = e^{-0.05x^{2}} \sin(x)$$


```python
x_values = np.linspace(-2*np.pi, 2*np.pi, 401)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x_values, func(x_values), label=f"${sp.latex(expr)}$")
plt.plot(x_values, deriv_func(x_values), label=f"dx/dy ${sp.latex(deriv)}$")

# plt.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")
plt.axhline(0, color='white',linewidth=0.5)
plt.axvline(0, color='white',linewidth=0.5)

plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.xlabel("x");
plt.ylabel("f(x)")
plt.legend();
plt.show();
```

- ### Podemos ver como claramente hay un mínimo global alrededor de x=-1.8
- ### Se aprecia un mínimo global alrededor de x=4.2

### Comencemos poniendo en marcha el GD

```python
# random starting point
local_min = np.random.choice(x_values,1)#np.array([6])#
#valor aleatorio del conjunto que hemos elegido de x para la función
print(f"Initial Random Local Min {local_min}")

# learning parameters
learning_rate = .01
training_epochs = 1000

# run through training
for _ in range(training_epochs):
  grad = deriv_func(local_min)
  local_min = local_min - learning_rate*grad
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")


# plot the results
plt.figure(figsize=(10,6))
plt.plot(x_values, func(x_values),'--') #f(x) original
plt.plot(x_values, deriv_func(x_values), '--') #f'(x)

#mínimo local encontrado por el GD, y su dericada
plt.plot(local_min, func(local_min),'ro')
plt.plot(local_min, deriv_func(local_min),'ro')

plt.axhline(0, color='cyan',linewidth=0.5)
plt.axvline(0, color='cyan',linewidth=0.5)
plt.xlim(x_values[[0,-1]])
plt.grid(color="gray", linestyle="--", linewidth=0.7)
#plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%local_min[0])
plt.show()
```

- ### En este ejemplo particular, el GD comenzó aleatoriamente en x=1.82, y logró alcanzar el mínimo global, que se encontraba a su derecha (pendiente negativa, avanza hacia la derecha). En este caso, el algoritmo se encontraba relativamente cerca de dicho punto

```python
local_min = np.random.choice(x_values,1)#np.array([6])#
#valor aleatorio del conjunto que hemos elegido de x para la función
print(f"Initial Random Local Min {local_min}")

# learning parameters
learning_rate = .01
training_epochs = 1000

# run through training
for _ in range(training_epochs):
  grad = deriv_func(local_min)
  local_min = local_min - learning_rate*grad
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")


# plot the results
plt.figure(figsize=(10,6))
plt.plot(x_values, func(x_values),'--') #f(x) original
plt.plot(x_values, deriv_func(x_values), '--') #f'(x)

#mínimo local encontrado por el GD, y su dericada
plt.plot(local_min, func(local_min),'ro')
plt.plot(local_min, deriv_func(local_min),'ro')

plt.axhline(0, color='cyan',linewidth=0.5)
plt.axvline(0, color='cyan',linewidth=0.5)
plt.xlim(x_values[[0,-1]])
plt.grid(color="gray", linestyle="--", linewidth=0.7)
#plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%local_min[0])
plt.show()
```

- ### En este otro, el GD comenzó en x=-4.9, y acabó en el punto x=6, que se encuentra e incluso fuera de la gráfica que habíamos trazado, debido a que el punto donde comenzó tenía una pendiente negativa, entonces se movió hacia la izquierda hasta desembocar en un "mínimo"

- ### Y en este otro, comenzó muy cerca del mínimo global, por lo que el algoritmo, al tener derivada negativa (la función se encuentra decreciendo en el punto donde apareció aleatoriamente), por lo que bajó hasta llegar al mínimo local.

## Ejecutando los Experimentos Paramétricos

___

### **Experimento 1:** Variar sistemáticamente las localizaciones ininciales

```python
#crearemos una función para visualizar gráficamente donde acabó cada uno de los GD de prueba

def plot_range(low, high):
    
    plt.plot(x_values, func(x_values), label=f"f(x) = ${sp.latex(expr)}$")
    plt.axvline(low, color='red',linewidth=1, ls="--")
    plt.axvline(high, color='red',linewidth=1, ls="--")


    # plt.plot(x_values, deriv_func(x_values), label=f"dx/dy ${sp.latex(deriv)}$")

    # plt.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")
    plt.axhline(0, color='white',linewidth=0.5)
    plt.axvline(0, color='white',linewidth=0.5)

    plt.grid(color="gray", linestyle="--", linewidth=0.5)
    plt.xlabel("x");
    plt.ylabel("f(x)")
    plt.legend();
    plt.show();
    

def plot_fixed_starting_loc(x_val):
    
    plt.plot(x_values, func(x_values), label=f"f(x) = ${sp.latex(expr)}$")
    plt.scatter(x_val, func(x_val), label=f"Fixed Starting Location on x={x_val}", color="red")


    # plt.plot(x_values, deriv_func(x_values), label=f"dx/dy ${sp.latex(deriv)}$")

    # plt.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")
    plt.axhline(0, color='white',linewidth=0.5)
    plt.axvline(0, color='white',linewidth=0.5)

    plt.grid(color="gray", linestyle="--", linewidth=0.5)
    plt.xlabel("x");
    plt.ylabel("f(x)")
    plt.legend();
    plt.show();
```

```python
# learning parameters
learning_rate = .01
training_epochs = 1000
```

```python
start_locs = np.linspace(-5,5,50)
final_res = np.zeros(len(start_locs))

# loop over starting points
for idx, local_min in enumerate(start_locs):
  
  # run through training
  for i in range(training_epochs):
    grad = deriv_func(local_min)
    local_min = local_min - learning_rate*grad
  
  # store the final guess
  final_res[idx] = local_min


# plot the results
plt.plot(start_locs,final_res,'s-')
plt.xlabel('Starting guess')
plt.ylabel('Final guess')
plt.show()
```

### - En el eje x de esta gráfica, tenemos cada punto x inincial que hemos decidido probar, y en el eje y el punto final a donde llegó

> - ### Podemos ver que en las primeras pruebas, donde el GD comenzó en valores de entre -5 y -4.2, el algoritmo desembocó en puntos cercanos.

```python
plot_range(-5,-4.2)
```

> - ### Desde el punto inincial x=-4 hasta x=2 aproximadamente, el GD finalizó en el mismo mínimo local (alrededor de x=-1)

```python
plot_range(-4,2) #el GD se moverá "cuesta abajo" por lo que en todo ese rango de x, el algoritmo desembocará en dicho mínimo global
```

> - ### Y desde x=2 hasta x=5, el GD también desembocó en un mínimo local común para todo este intervalo (alrededor de x=4)

```python
plot_range(2,5)
```

### **Experimento 2:** Variar sistemáticamente las tasas de aprendizaje (`learning rates`). (forzando el valor x inincial a x=0)

```python
plot_fixed_starting_loc(x_val=0)
```

### Para este experimento, mantendremos "constante" o "predefinido" el valor inincial, que será x=0. Por lo que, podemos saber a priori cuál será el lugar deseable donde debería acabar el GD (alrededor de -1.5, mínimo global, a la izquierda) 

```python
learning_rates = np.linspace(1e-10,1e-1,50) #Definimos un vector de 50 combinaciones de tasas de aprendizaje que vayan desde 1e-10 hasta 1e-1 (0.1)
final_res = np.zeros(len(learning_rates)) #otro vector de ceros inincialmente para almacenar los resultados de cada prueba de GD

# loop over learning rates
for idx,learning_rate in enumerate(learning_rates):
  
  # force starting guess to 0
  local_min = 0 #de esta manera, podremos correctamente comprobar cómo fue el desempeño de cada GD sin valores aleatorios

  # run through training
  for i in range(training_epochs):
    grad = deriv_func(local_min)
    local_min = local_min - learning_rate*grad
  
  # store the final guess
  final_res[idx] = local_min


plt.plot(learning_rates,final_res,'s-')
plt.xlabel('Learning rate')
plt.ylabel('Final guess')
plt.show()
```

### - Para valores pequeños de `learning_rate`, (hasta 0), podemos ver que el GD terminó alrededor del mismo punto donde comenzó, debido a que el gradiente se vuelve tan pequeño que apenas se mueve el valor de x

```python
2 - (-3*1e-10) #un ejemplo de este hecho, valor ejemplo x=2, derivada -3, resultando en una actualización de x mínimamente apreciable
```

### - A partir de 0, comienza a ir desembocando en x=-1.2 o x=-1.4 (mínimo deseado)

### - Podemos ver cómo en el rango de `learning_rates` (desde 0.005 aproximadamente hasta 0.1), el mínimo donde acaba el GD es igual para todos estos valores, por lo que en casos de datasets, este seria un caso donde el modelo aprende de forma robusta los datos en un rango considerable

> - ### Como detalle, existen algoritmos como "Adam" que pueden ajustar el `learning_rate` durante el tiempo

### Visto lo anterior, para valores de `learning_rate` bajos, como el GD tomará pasos más "pequeños", entonces necesitaríamos ejecutarlo durante más `training_epochs`, pudiendo llegar a buenos resultados. Esto nos lleva a:

__________

### **Experimento 3:** Analizar la interacción entre el valor del `learning_rate` y el número de `training_epochs`. (Con x=0 valor inincial) 

```python
plot_fixed_starting_loc(x_val=0)
```

### De la misma manera, comenzamos con un valor inincial de x predefinido (x=0), por lo que el mínimo deseable a donde debería el GD desembocar sería alrededor de -1.4 o -1.5

```python
# Experiment 3: interaction between learning rate and training epochs

# setup parameters
learning_rates = np.linspace(1e-10,1e-1,50)
training_epochs = np.round(np.linspace(10,500,40))

# initialize matrix to store results
final_res = np.zeros((len(learning_rates),len(training_epochs))) #como ahora analizamos 2 variables, usamos una matriz (2 columnas)



# loop over learning rates
for lr_idx,learning_rate in enumerate(learning_rates): #para cada valor distinto de learning_rate ejecutamos el GD un valor distinto de epochs

  # loop over training epochs
  for ep_idx,train_epochs in enumerate(training_epochs):
  
    # run through training (again fixing starting location)
    local_min = 0
    for i in range(int(train_epochs)):
      grad = deriv_func(local_min)
      local_min = local_min - learning_rate*grad
    
    # store the final guess
    final_res[lr_idx, ep_idx] = local_min
```

```python
# plot the results

fig,ax = plt.subplots(figsize=(7,5))

plt.imshow(final_res,extent=[learning_rates[0],learning_rates[-1],
                             training_epochs[0],training_epochs[-1]], #trazamos la matriz configurando el tamaño 
           aspect='auto',origin='lower',vmin=-1.45,vmax=-1.2)

plt.xlabel('Learning rate')
plt.ylabel('Training epochs')
plt.title('Final guess')
plt.colorbar()
plt.show()
```

- ### En esta gráfica de la matriz, podemos ver cómo relacionamos en el eje x (valores probados de `learning_rate`) y en el eje y (valores probados de `training_epochs`)

- ### Cada "píxel" es un experimento (combinación de los 2 parámetros) diferente

- ### Para cada experimento, tenemos una escala de color, donde colores más "morados" corresponden a ejecuciones del GD cercanas al mínimo deseado (x=-1.45), mientras que valores más "amarillentos" corresponden a ejecuciones del GD donde acabó en zonas menos "bajas", o a "medio camino", siendo "peores soluciones"

- ### Por tanto, vemos que en este caso, para valores "grandes" de `learning_rate` (superiores a 0.02) y un número alto de `training_epochs` (superior a 100), obtendremos un resultado considerable (es decir, el GD acaba en el mínimo deseado)

```python
# another visualization
plt.plot(learning_rates,final_res)
plt.xlabel('Learning rates')
plt.ylabel('Final function estimate')
plt.title('Each line is a training epochs N')
plt.show()
```

- ### En esta otra gráfica, visualizamos en el eje x los distintos valores de `learning_rate` y en el eje y, el valor final del GD

- ### Por lo que, líneas (pruebas de GD con N epochs) que lleguen hasta el fondo de la gráfica (eje y -1.4), serán combinaciones de ese valor de `learning_rate` (eje x) y `training_epochs` óptimas para este caso particular
