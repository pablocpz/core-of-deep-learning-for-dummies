## Wine Quality Dataset

```python
import torch
from torch.utils.data import DataLoader,TensorDataset
from sklearn.model_selection import train_test_split
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import seaborn as sns
```

Vemos variables independientes altamente correlacionadas, como son la caracterstica `fixed acidity`, que tomando valores ms altos tambin hace incrementar en valores ms altos de `citric acid`

```python
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

Vemos la diversidad de rangos de valores que toma cada variable, por ejemplo, `total sulfur dioxide` toma valores ms amplios y grandes

- Deberemos escalar a un mismo rango las caractersticas, para un mejor desempeo del modelo de DL

- Adems, podemos ver valores atpicos **(outliers)** por ejemplo en `total sulfur dioxide`, tomando valores de en torno a 300, lo que no corresponde a su distribucin normal

```python
data = data[data["total sulfur dioxide"]<200] #eliminamos esos dos outliers de forma manual
```

```python
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

Para escalar las variables, usaremos el `z-score`

```python
#obtenemos todas las columnas a normalizar (variables independientes)
cols_zscore = data.keys()
cols_zscore = cols_zscore.drop("quality")

for col in cols_zscore:
    
    mean_val = np.mean(data[col])
    st_dev = np.std(data[col], ddof=1)
    data[col] = (data[col]-mean_val) / st_dev
    
#tambin puede hacerse as
# data[cols_zscore] = data[cols_zscore].apply(stats.zscore)
```

Media alrededor de 0, y desviacin estandar=1

```python
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

Las caractersticas ahora se sitan en torno a un rango similar, excepto la variable a predecir

```python
plt.title("Quality Rating Histogram")
plt.ylabel("Count")
data.quality.hist(color="aqua");
```

Hay un gran desbalance. Vemos que valores de valoracin de 5 y 6 estn presentes en los datos demasiado en comparacin con otros valores

> Para ello, y evitar este imbalance en los datos para el modelo, estableceremos la variable objetivo a ser 0s o 1s, (si la valoracin es > 5)

```python
pd.options.mode.chained_assignment = None
```

```python
data['bool_quality'] = 0
data["bool_quality"][data['quality']>5] = 1
```

```python
data[["quality", "bool_quality"]].head()
```

## Convitiendo los datos a tensores

```python
X = torch.tensor(data[cols_zscore].values).float()
y = torch.tensor(data["bool_quality"].values).float()
```

```python
print(X.shape)
print(y.shape)
```

Convertimos las etiquetas a formato multidimensional

```python
y = y.reshape(-1,1)
#y = y[:,None]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)

batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size,
                         shuffle=True, drop_last=True)
#drop_last se asegura de quitar el ltimo batch con tamao diferente

test_loader = DataLoader(test_dataset, batch_size=test_dataset.tensors[0].shape[0])
```
