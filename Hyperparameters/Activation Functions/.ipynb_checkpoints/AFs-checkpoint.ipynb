## Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4, 100)
```

# Sigmoid

$\large{\sigma(x) = \frac{1}{1+ e^{-x}}}\\ $

$ \large{\sigma(x)' = \sigma(x) \cdot (1 - \sigma(x))}$

```python
def sigmoid(x):

  return 1/(1+np.e**(-x))
```

```python
def sigmoid_deriv(x):

  #f'(x)
  return sigmoid(x)*(1-sigmoid(x))
```

```python
plt.style.use("default")
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, sigmoid(x), label=r'Sigmoid $\sigma(x) = \frac{1}{1+ e^{-x}}$');
plt.plot(x, sigmoid_deriv(x), label=r"$\sigma(x)' = \sigma(x) \cdot (1 - \sigma(x))$");

plt.grid(alpha=0.4)
plt.legend(fontsize=15)
plt.show();
```

# Tanh
$\large{\phi(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}}\\ $

$ \large{\phi(x)' = 1 - \phi(x)^2}$

```python
def tanh(x):

  return (np.e**x - np.e**-x) / (np.e**x + np.e**-x)
  # return 2/(1+np.e**(-2*x)) - 1

def tanh_deriv(x):

  return 1-(tanh(x)**2)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, tanh(x), label=r'Tanh $\phi(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$');
plt.plot(x, tanh_deriv(x), label=r"$\phi(x)' = 1 - \phi(x)^2$");

plt.grid(alpha=0.4)
plt.legend(fontsize=15)
plt.show();
```

# ReLU

$\large{\omega(x) = \max(0,x)} = \begin{cases}
      0, & \text{if } x < 0 \\
      x, & \text{otherwise}
   \end{cases}\\
\large{\omega(x)' = \begin{cases}
      0, & x < 0 \\
      1, & x > 0
   \end{cases}} \\
\text{(Undifferentiable at x=0)} \\
$

```python
def relu(x):
  return np.maximum(0,x)

def relu_deriv(x):

  return [0 if x_i < 0 else 1 for x_i in x]
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, relu(x), label=r'ReLU $\omega(x) = \max(0,x)}$');
plt.plot(x, relu_deriv(x), label=r"$\omega(x)' =$ 0 if x<0 else 1");

plt.grid(alpha=0.4)
plt.legend(fontsize=15)
plt.show();
```

## ReLU Variants
https://arxiv.org/pdf/2109.14545.pdf

> ### Leaky ReLU

$\large{LReLU(x) = \max(\underline{\alpha} x,x)} = \begin{cases}
      \underline{\alpha} \cdot x, & x < 0 \\
      x, & \text{otherwise}
   \end{cases}\\
\large{LReLU(x)' = \begin{cases}
      \underline{\alpha}, & x < 0 \\
      1, & x > 0
   \end{cases}} \\
\large{\text{Where } \alpha \text{ is a parameter scaling factor for negative input, typically } 0 < \alpha < 1}
$

```python
def leaky_relu(x, alpha):
  return np.maximum(alpha*x, x)

def leaky_relu_deriv(x, alpha):
  return [alpha if x_i < 0 else 1 for x_i in x]
```

```python
leaky_relu(-50000, alpha=0.01)
```

```python
plt.figure(figsize=(10,6))
plt.title(r"$\alpha = 0.1$ (negative slope)")
plt.plot(x, leaky_relu(x,0.1), label=r'$LReLU(x) = \max(x \cdot \alpha,x)}$');
plt.plot(x, leaky_relu_deriv(x,0.1), label=r"$LReLU(x)'= \alpha,$if x<0 else 1");

plt.grid(alpha=0.4)
plt.legend(fontsize=15)
plt.show();
```
