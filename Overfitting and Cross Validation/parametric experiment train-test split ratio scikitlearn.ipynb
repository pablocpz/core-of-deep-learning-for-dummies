## Creando el la función que crea el modelo

```python
X_train.shape[1]#tenemos 4 inputs de entrada
```

```python
len(y_train.unique()) #3 categorías distintas (etiquetas), 0,1 y 2 respectivamente
```

```python
class Model(nn.Module):
    
    
    def __init__(self):
        super().__init__()
        
        self.input = nn.Linear(X_train.shape[1], 64) #capa de entrada (1 capa oculta de 64 nodos)
        self.hid = nn.Linear(64,64) #2da capa oculta
        self.out = nn.Linear(64,len(y_train.unique())) #capa de salida   
        
    def forward(self, x):
        
        x = F.relu(self.input(x))
        x = F.relu(self.hid(x))
        x = self.out(x)
        # la función de pérdida CrossEntropyLoss() implementa en la misma clase la función Softmax de salida
        return x    
```

```python
def build_model():
    
    model = Model()
    loss_func = nn.CrossEntropyLoss() #ver docs (usa LogSoftmax debido a la propiedad de los logaritmos de hacer los números pequeños "más espaciados")
    #por lo que añade el logaritmo a la función softmax

    learning_rate = 0.01

    optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
    
    return model, loss_func, optim
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Creando la función que entrena el modelo

```python
n_epochs = 200
```

### Fíjate que no se da overfitting, ya que el conjunto de testing no está influyendo sobre el entrenamiento, tan solo sirve para almacenar los datos del experimento

```python
def train(training_ratio):
    
    #inincializamos las pérdidas
    losses = torch.zeros(n_epochs)
    
    train_accs = []
    test_accs = []
    #si la división de los datos se diese dentro de las epochs sería overfitting
    #ya que es probable que, al ser divisiones aleatorias, lo que era antes de testing fuese ahora de training, haciendo que dicho set de test sea usado por el modelo para entrenarse
    
    
    #separamos los datos en base al ratio del parámetro
    #en este caso, especificamos el argumento train_size
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                        train_size=training_ratio,)
    X_train = torch.tensor(X_train).float() #matriz de características
    y_train = torch.tensor(y_train).long() #vector de las etiquetas transformadas a números (tipo de dato long() -> entero) #podiamos tambien haber hecho df.iris[df.iris.species=="setosa"] == 1 y así

    X_test = torch.tensor(X_test).float() #matriz de características
    y_test = torch.tensor(y_test).long()     
    
    for epoch in range(n_epochs): #tqdm
        
        
        #forward pass
        y_hat = model(X_train)

        #compute error
        loss = loss_func(y_hat, y_train)
        losses[epoch] = loss

        #backward pass
        optim.zero_grad() #clear previous gradients cache
        loss.backward() #compute gradients based on the error
        optim.step() #update parameters

        #calculamos la precisión
       
        train_acc = 100*torch.mean(((torch.argmax(y_hat, axis=1) == y_train).float()))
        test_acc = 100*torch.mean(((torch.argmax(model(X_test), axis=1) == y_test).float()))
        
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        
    return train_accs, test_accs
```

_________

## Testeando el modelo una sola vez

```python
model, loss_func, optim = build_model()

train_accs, test_accs = train(training_ratio=0.8)
```

```python
train_accs[:3]
```

```python
plt.figure(figsize=(10,6))

plt.plot(train_accs, "mo-")
plt.plot(test_accs, "yo-")
plt.xlabel("No. Of Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Test"])
plt.show();
```

____________

## Hora del experimento!

```python
n_epochs = 200

train_sizes = np.linspace(0.2, 0.95, 10)

total_train_accs = np.zeros((len(train_sizes), n_epochs))
total_test_accs = np.zeros((len(train_sizes), n_epochs))

for idx in range(len(train_sizes)):
    
    
    model, loss_func, optim = build_model()
    
    train_accs, test_accs = train(training_ratio=train_sizes[idx])
    
    #almacenamos los resultados
    total_train_accs[idx,:] = train_accs
    total_test_accs[idx,:] = test_accs
```

_____

## Graficando los resultados del experimento

```python
fig, axs = plt.subplots(nrows=1, ncols=2,
                       figsize=(17,5), sharex=True, sharey=True)


axs[0].imshow(total_train_accs, aspect="auto",
             vmin=50, vmax=90, extent=[0, n_epochs, train_sizes[-1], train_sizes[0]])

axs[0].set_xlabel("No. Of Epoch")
axs[0].set_ylabel("Training Ratio")
axs[0].set_title("Training Accuracy")

im = axs[1].imshow(total_test_accs, aspect="auto",
             vmin=50, vmax=90, extent=[0, n_epochs, train_sizes[-1], train_sizes[0]])

axs[1].set_xlabel("No. Of Epoch")
axs[1].set_ylabel("Training Ratio")
axs[1].set_title("Test Accuracy")
fig.colorbar(im, ax=axs[1])
plt.show()
```
