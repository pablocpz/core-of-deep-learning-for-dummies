```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
plt.style.use("dark_background")
```

```python
x = sp.symbols("x")
```

```python
expr = 3*x**2 -3*x +4
func = sp.lambdify(x, expr, "numpy")
```

```python
expr
```

```python
deriv = sp.diff(expr)
deriv_func = sp.lambdify(x, deriv, "numpy")
```

```python
deriv
```

### Visualizamos la relación entre la función original y su primera derivada

```python
x_values = np.linspace(-2,2,2001)
plt.plot(x_values, func(x_values), label=f"${sp.latex(expr)}$")
plt.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")
plt.axhline(0, color='white',linewidth=0.5)
plt.axvline(0, color='white',linewidth=0.5)

plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.xlabel("x");
plt.ylabel("f(x)")
plt.legend();
plt.show();
```

### Cuando la función f(x) original decrece, f'(x) será negativa, y viceversa

```python
local_min = np.random.choice(x_values,1)
print(f"Initial Random Local Min {local_min}")

learning_rate = .01
training_epochs = 100

for _ in range(0, training_epochs):
    grad = deriv_func(local_min)
    local_min = local_min - (grad*learning_rate)
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")
```

```python
#storing the model gradient (derivative) and local min at each epoch

local_min = np.random.choice(x_values,1)
print(f"Initial Random Local Min {local_min}")

learning_rate = .01
training_epochs = 100

model_params = np.zeros((training_epochs, 2)) #shape epochs x 2 cols

for i in range(0, training_epochs):
    grad = deriv_func(local_min)
    local_min = local_min - (grad*learning_rate)
    model_params[i,0] = local_min[0]
    model_params[i,1] = grad[0]     
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(model_params[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {local_min[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

#### - En el primer subplot se aprecia que el mínimo local comienza aleatoriamente en x=0.15, y con el paso de las iteraciones (proceso de "aprendizaje"), va acercándose hasta 0.50 que es el verdadero mínimo de f(x) original

#### - La derivada del mínimo local en cada iteración va decreciendo, buscando "ir hacia abajo" y estacionarse en un lugar donde la función f(x) se encuentre con pendiente nula (0) (vemos cómo se aproxima de forma "asintótica")

#### - Podemos apreciar que la derivada comienza siendo negativa, significando que el punto aleatorio que comenzó como mínimo local (0.15) se encontraba decreciendo

#### - A partir de alrededor de 70 iteraciones en ambos casos, tanto la derivada (gradiente) como el mínimo local se comienzan a estacionarse (es decir, el algoritmo de Gradiente Descendiente en 1D ya habría convergido)

```python
#storing the model gradient (derivative) and local min at each epoch

local_min = np.random.choice(x_values,1)
print(f"Initial Random Local Min {local_min}")

learning_rate = .01
training_epochs = 100

model_params = np.zeros((training_epochs, 2)) #shape epochs x 2 cols

for i in range(0, training_epochs):
    grad = deriv_func(local_min)
    local_min = local_min - (grad*learning_rate)
    model_params[i,0] = local_min[0]
    model_params[i,1] = grad[0]     
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(model_params[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {local_min[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

- #### Ahora, de forma inversa, el mínimo local aleatorio comenzó en un punto positivo, significando que el gradiente debe ir haciendo decrecer ese valor de x buscando la derivada cercana a 0 para converger

- #### De esta forma, podemos ver cómo depende del mínimo aleatorio que se inincialize, que el algoritmo tendrá más o menos ventaja dependiendo de lo cerca o alejado que se encuentre de posibles soluciones (como x=0.5 en este caso)

-------------------------------------

#### Ahora, probemos a variar el `learning_rate` y las `training epochs`

##### Recordando que:

- #### El `learning_rate` ajusta la longitud o "cómo de grandes" son los pasos que toma el gradiente para avanzar a lo largo del dominio de la función en busca de ese posible mínimo *debido a que, si calculas el nuevo mínimo: `antiguo mínimo - (su derivada*learning_rate)`), valores grandes de learning_rate provocarán que se sume/reste (dependiendo si el mínimo es positivo/negativo), que aumente/disminuya más rápido o más lento dicho valor de x*



```python
#imagina: el mínimo actual es 34, si el lr es de 0.0001 (número más pequeño) -> decrece (al ser positivo) más lento
34 - (-4*0.0001) #imaginamos que la derivada de x=34 es -4
```

```python
#pero si el lr es 0.1 (más grande), decrece x a pasos más grandes
34 - (-4*0.1) #imaginamos que la derivada de x=34 es -4
```

```python
def grad_1D(learning_rate, training_epochs=100, visualize=False):
    
    local_min = np.random.choice(x_values,1)
    print(f"Initial Random Local Min {local_min}")

    model_params = np.zeros((training_epochs, 2)) #shape epochs x 2 cols

    for i in range(0, training_epochs):
        grad = deriv_func(local_min)
        local_min = local_min - (grad*learning_rate)
        model_params[i,0] = local_min[0]
        model_params[i,1] = grad[0]     

    print(f"Final Local Min after {training_epochs} epochs: {local_min}")

    if visualize:
        fig,ax = plt.subplots(1,2,figsize=(12,4))

        for i in range(2):
          ax[i].plot(model_params[:,i],'o-')
          ax[i].set_xlabel('Iteration')
          ax[i].set_title(f'Final estimated minimum: {local_min[0]:.5f}')

        ax[0].set_ylabel('Local minimum')
        ax[1].set_ylabel('Derivative')

        plt.subplots_adjust(hspace=0.5)  # Ajustar el espaciado vertical entre los subplots


        plt.show();
```

```python
grad_1D(learning_rate=1)
```

* ### Vemos que si establecemos un `learning_rate` demasiado grande el gradiente avanzará a pasos tan "largos" que saltará los mínimos locales, llegando a números o bien muy grandes (ejemplo 7.57306469e+69) o muy pequeños 7.57306469e+69

```python
grad_1D(learning_rate=0.001, visualize=True)
```

- #### En este caso, apreciamos, que con un `learning_rate=0.01`, el algoritmo no consigue converger, ya que el mínimo local va cambiando de forma elevada, y no se aproxima ni se "estaciona" en torno a ningún valor de x, por lo tanto, `100 epochs` no serían suficientes en este caso

```python
from IPython.display import HTML
from matplotlib import animation
```

```python
def grad_1D(expr, x_values, learning_rate, training_epochs, display_animation=False):
    
    func = sp.lambdify(x, expr, "numpy")
    deriv = sp.diff(expr)
    deriv_func = sp.lambdify(x, deriv, "numpy")
    
    #algorithm
    local_min = np.random.choice(x_values,1)
    initial_local_min = local_min
    print(f"Initial Random Local Min {initial_local_min}")
    model_params = np.zeros((training_epochs, 2)) #shape epochs x 2 cols
    for i in range(0, training_epochs):
        grad = deriv_func(local_min)
        local_min = local_min - (grad*learning_rate)
        model_params[i,0] = local_min[0]
        model_params[i,1] = grad[0]     
    print(f"Final Local Min after {training_epochs} epochs: {local_min}")
    
    #visualize algorithms details per iteration
    fig,ax = plt.subplots(1,2,figsize=(12,4))
    for i in range(2):
      ax[i].plot(model_params[:,i],'o-')
      ax[i].set_xlabel('Iteration')
      ax[i].set_title(f'Final estimated minimum: {local_min[0]:.5f}')

    ax[0].set_ylabel('Local minimum')
    ax[1].set_ylabel('Derivative')
    plt.show()
    
    if display_animation:
        #prepare animation
        grad_fig, ax = plt.subplots(figsize=(12,6), dpi=100)

        ax.plot(x_values, func(x_values), label=f"${sp.latex(expr)}$")
        ax.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")

        plt.title(f"Empirical Local Minimum: {local_min[0]}")


        plt.axhline(0, color='white',linewidth=0.5)
        plt.axvline(0, color='white',linewidth=0.5)

        plt.grid(color="gray", linestyle="--", linewidth=0.5)
        plt.xlabel("x");
        plt.ylabel("f(x)")
        plt.legend();
        plt.close()

        def tangent_line(x, x1, y1):
            # m*x+b
            return deriv_func(x1)*(x-x1) + y1

        title = ax.set_title('', fontweight="bold")
        local_min_scat = ax.scatter(initial_local_min, func(initial_local_min), color="orange")
        initial_tangent_range = np.linspace(initial_local_min-0.5, initial_local_min+0.5, 10)
        tangent_plot = ax.plot(initial_tangent_range, tangent_line(x=initial_tangent_range,
                                                                   x1=initial_local_min,
                                                                   y1=func(initial_local_min)), linestyle="--",  color="orange", linewidth=2)[0]
        grad_annotation = ax.annotate(
            'Gradient={0:2f}'.format(deriv_func(initial_local_min[0])),
            xy=(initial_local_min,func(initial_local_min)), xytext=(initial_local_min,func(initial_local_min)+1),
            arrowprops = {'arrowstyle': "-", 'facecolor' : 'orange'},
            textcoords='data', color='orange' , rotation=20, fontweight="bold"
        )

        def drawframe(epoch):
            title.set_text('Epoch={0:4d}'.format(epoch, model_params[epoch, 1]))
            x1 = model_params[epoch, 0]
            y1 = func(model_params[epoch, 0])
            local_min_scat.set_offsets((x1, y1))
            tangent_range = np.linspace(x1-0.5, x1+0.5, 10)
            tangent_values = tangent_line(x=tangent_range, x1=x1 ,y1=y1)
            tangent_plot.set_xdata(tangent_range)
            tangent_plot.set_ydata(tangent_values)
            grad_annotation.set_position((x1, y1+1))
            grad_annotation.xy = (x1, y1)
            grad_annotation.set_text('Gradient={0:2f}'.format(model_params[epoch, 1]))
            return local_min_scat,

        # blit=True re-draws only the parts that have changed.
  
        anim = animation.FuncAnimation(grad_fig, drawframe, frames=training_epochs, repeat=False, interval=500, blit=True) 


        writer = animation.PillowWriter(fps=30,
                                        metadata=dict(artist='Me'),
                                        bitrate=1800)
        # ani.save('gradient1D.gif', writer=writer)
        # from IPython.display import HTML
        # HTML(anim.to_html5_video())

        return anim, writer
```

```python
anim, writer = grad_1D(expr=expr, 
                       x_values=x_values, learning_rate=0.001, training_epochs=100, display_animation=True)
```

- #### Si probamos con un `learning_rate` más bajo, como un `0.001`, nos damos cuenta de lo mismo, parece que el algoritmo no tiene suficientes iteraciones para llegar a converger correctamente, provocando que en la iteración número 100, la derivada (gradiente) sea -2, por lo que claramente no ha convergido, ya que no es un valor cercano a 0

```python
from IPython.display import HTML
HTML(anim.to_html5_video())
```

- #### Con esta animación vemos claramente, cómo, teniendo un `learning_rate` pequeño, los pasos serán "pequeños" a lo largo del eje x, y por tanto necesitaremos más `training epochs`. Fíjate, que cuando acaba la iteración número 100, se ve cómo la tangente no es horizontal, es decir, la pendiente no es 0, el algoritmo no le dió tiempo a converger, quedándose a un valor alejado de x=0.5 (mínimo local)

```python
anim, writer = grad_1D(expr=expr,
                       x_values=x_values, learning_rate=0.001, training_epochs=1000, display_animation=True)
anim.save('gradient1D.gif', writer=writer)
from IPython.display import HTML
HTML(anim.to_html5_video())
```

- #### Ahora, con `1000 training epochs`, el gradiente descendente 1D obtiene un mejor resultado, a pesar de tener un `learning_rate` más bajo
> #### Por tanto, cuando el `learning_rate` es más bajo, necesitaremos más `training_epochs` para alcanzar un buen resultado, ya que avanza de forma más lenta.

```python
#storing the model gradient (derivative) and local min at each epoch

local_min = np.random.choice(x_values,1)
print(f"Initial Random Local Min {local_min}")

learning_rate = .01
threshold = 0.001

model_params = []

grad = np.Infinity

training_epochs = 0
while grad > threshold or grad < -threshold:
    
    grad = deriv_func(local_min)
    local_min = local_min - (grad*learning_rate)
    
    model_params.append([local_min[0], grad[0]])
    
    training_epochs+=1
    
print(f"Final Local Min after {training_epochs} epochs: {local_min}")

fig,ax = plt.subplots(1,2,figsize=(12,4))

model_params = np.array(model_params)

for i in range(2):
  ax[i].plot(model_params[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {local_min[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

#### Podemos ver, que llevará menos `training_epochs`, es decir, el algoritmo tardará menos en converger dependiendo de donde se inincialize el valor aleatorio de mínimo local (x). En este caso, al estar tan cerca del mínimo local de la función (x=0.5), entonces el algoritmo llegó al mínimo local en tan solo 42 `epochs`(usando un "límite" que detiene el algoritmo al llegar a un valor concreto de gradiente), llegando a la conclusión de que:


- #### Dependiendo de donde se inincialize el mínimo local aleatorio, el algoritmo del GD se beneficiará más o menos, suponiendo mayor o menor tiempo de cómputo, y de convergencia en sí
