## **Solución 1:** G.D. using a fixed `learning_rate`

### (Usando el GD standard, el learning rate fijo durante todas las training epochs)

_____

```python
# random starting point
localmin = np.random.choice(x_values,1)
initval = localmin[:] # store the initial value

# learning parameters
learning_rate = .01
training_epochs = 50

# run through training and store all the results
modelparamsFixed = np.zeros((training_epochs,3))
for i in range(training_epochs):
  
  # compute gradient
  grad = deriv_func(localmin)

  # non-adaptive learning rate
  lr = learning_rate

  # update parameter according to g.d.
  localmin = localmin - lr*grad

  # store the parameters
  modelparamsFixed[i,0] = localmin[0]
  modelparamsFixed[i,1] = grad[0]
  modelparamsFixed[i,2] = lr
```

## **Solución 2:** G.D. using a gradient-based `learning_rate`

______

### Usando esta técnica, el `learning_rate` estará escalado por la **magnitud del gradiente**. Es decir, si la función crece/decrece a una razón alta, entonces el learning_rate aumentará y conforme el GD se aproxime a un mínimo local, la magnitud del gradiente (el signo no importa, usamos valor absoluto) será cada vez más pequeño, haciendo decrecer también el learning_rate

- ### En definitiva, el valor del `learning_rate` se hace más grande si el GD se encuentra alejado del mínimo local, y se vuelve más pequeño conforme se acerca a dicho mínimo

```python
0.01 * np.abs(7) #imagina el valor actual de lr que sea 0.01, y la magnitud del gradiente es 7, por lo que el lr crecerá
```

```python
0.01 * np.abs(-1.5) #en cambio, si la magnitud del gradiente es pequeña, el lr también será cada vez más pequeño.
```

```python
# random starting point
localmin = np.random.choice(x_values,1)
initval = localmin[:] # store the initial value

# learning parameters
learning_rate = .01
training_epochs = 50

# run through training and store all the results
modelparamsGrad = np.zeros((training_epochs,3))
for i in range(training_epochs):
  
  # compute gradient
  grad = deriv_func(localmin)

  # adapt the learning rate according to the gradient
  lr = learning_rate*np.abs(grad)

  # update parameter according to g.d.
  localmin = localmin - lr*grad

  # store the parameters
  modelparamsGrad[i,0] = localmin[0]
  modelparamsGrad[i,1] = grad[0]
  modelparamsGrad[i,2] = lr[0]
```

## **Solución 3:** G.D. using a time-based `learning_rate`

- ### Con este enfoque, el nuevo `learning_rate` = valor_actual*(1 - (epoch_actual+1)/training_epochs)

```python
(0+1)/50 #ejemplo: comienza el lr siendo 0.1. En la epoch 0:
```

```python
1 - (0+1)/50 #nos dará un valor cercano a 1, ya que se lo restamos
```

```python
0.1*(1 - (0+1)/50) #y se lo multiplicamos, haciendo el nuevo lr más pequeño con el paso de las epochs
```

```python
0.098*(1 - (1+1)/50) #pasa a la epoch 1, ahora el valor del lr es más bajo aún
```

- ### Podemos ver que, conforme pasen las epochs de entrenamiento, el `learning_rate` va decreciendo también

```python
# redefine parameters
learning_rate = .1
localmin = initval

# run through training and store all the results
modelparamsTime = np.zeros((training_epochs,3))
for i in range(training_epochs):
  grad = deriv_func(localmin)
  lr = learning_rate*(1-(i+1)/training_epochs)
  localmin = localmin - lr*grad
  modelparamsTime[i,0] = localmin[0]
  modelparamsTime[i,1] = grad[0]
  modelparamsTime[i,2] = lr
```

```python
def plot_point(x_val):
    plt.figure(figsize=(4,2))
    plt.plot(x_values, func(x_values), label=f"f(x) = ${sp.latex(expr)}$")
    plt.scatter(x_val, func(x_val), label=f"x={x_val}", color="red")


    # plt.plot(x_values, deriv_func(x_values), label=f"dx/dy ${sp.latex(deriv)}$")

    # plt.plot(x_values, deriv_func(x_values), label=f"dy/dx ${sp.latex(deriv)}$")
    plt.axhline(0, color='white',linewidth=0.5)
    plt.axvline(0, color='white',linewidth=0.5)

    plt.grid(color="gray", linestyle="--", linewidth=0.5)
    plt.xlabel("x");
    plt.ylabel("f(x)")
    plt.legend();
    plt.show();
```

```python
fig,ax = plt.subplots(1,3,figsize=(10,3))

# generate the plots
for i in range(3):
  ax[i].plot(modelparamsFixed[:,i],'o-',markerfacecolor='w', color="red")
  ax[i].plot(modelparamsGrad[:,i],'o-',markerfacecolor='w', color="orange")
  ax[i].plot(modelparamsTime[:,i],'o-',markerfacecolor='w')
  ax[i].set_xlabel('Iteration')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')
ax[2].set_ylabel('Learning rate')
ax[2].legend(['Fixed l.r.','Grad-based l.r.','Time-based l.r.'])

plt.suptitle("Learning Rate Modifications Comparison (random starting location)")

plt.tight_layout()
plt.show()
```

### Tenemos tres gráficos, en cada uno de ellos muestra las `training_epochs` en el eje x, y el mínimo local, derivada, y `learning_rate` en el eje y, respectivamente **(el valor inincial del GD de cada uno de los 3 "tipos de GD" es aleatorio)**:

> - ### Para el **GD con lr fijo**, el mínimo local va decreciendo hasta llegar a un valor de x en torno a 0.5 (mínimo deseado), al cabo de 50 epochs. Su derivada, también va decreciendo, como es lógico hasta acabar en el mínimo. Y el learning rate es constante, ya que no se cambia

> - ### Para el **GD con lr basado en gradiente**, el mínimo local va aproximándose hasta x=0.5 hasta llegar a alcanzarlo alrededor de las 20 epochs. Su derivada comienza siendo negativa, por lo que va decreciendo hasta acabar en 0, y el lr va decreciendo a causa del gradiente

> - ### Para el **GD con lr basado en tiempo**, consigue llegar a x=0.5 en apenas 5 epochs, por lo que la derivada se estaciona alrededor de 0 también bastante rápido, y el learning rate, al estar escalado por las epochs, con el paso de las epochs va decreciendo de forma lineal hasta ser cercano a 0).

### **De modo que, para este caso particular, el `learning rate basado en epochs` fue el más rápido y eficiente**
