## Experiment

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from IPython import display
# %config InlineBackend.figure_formats = ['svg']
%matplotlib inline
# plt.style.use("dark_background")
```

```python
n_per_clust = 100 #número de datos que habrán alrededor de cada centro de categoría
blur = 2 #std, indica cómo de "dispersos" estarán los datos

#coordenadas de la nube de datos de cada categoría (solo eje x)
A = [  1, 3 ] #el centro de la categoría A (x=1,y=1)
B = [  1, -2 ] #el centro de la categoría B (x=5,y=1)

# generar datos aleatorios o "ruido" alrededor de cada centro (A,B)
a = [ A[0]+np.random.randn(n_per_clust)*blur , A[1]+np.random.randn(n_per_clust)*blur ]
b = [ B[0]+np.random.randn(n_per_clust)*blur , B[1]+np.random.randn(n_per_clust)*blur ]

# true labels
#para las etiquetas, asignamos ceros a la categoría A, y 1 a la categoría B
labels_np = np.vstack((np.zeros((n_per_clust,1)),np.ones((n_per_clust,1))))

# concatenamos los datos aleatorios en una matriz traspuesta
data_np = np.hstack((a,b)).T

# convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# show the data
fig = plt.figure(figsize=(10,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'ko')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'go')
plt.title('The qwerties!')
plt.xlabel('qwerty dimension 1')
plt.ylabel('qwerty dimension 2')
plt.show()
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
import scipy.stats as stats
#data scaling (z-score)

X = stats.zscore(data)
```

```python
from sklearn.model_selection import train_test_split
y = labels
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

- Convertir a tensor pytorch

- Crear TensorDataset

- Convertirlos a dataloaders de batches

```python
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
```

```python
batch_size = 16
```

```python
train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                          drop_last=True)
test_loader = DataLoader(train_dataset, batch_size=test_dataset.tensors[0].shape[0])
```

```python
next(iter(train_loader))
```

```python
train_dataset.tensors[0].shape
```

```python
class Model(nn.Module):

  def __init__(self):

    super().__init__()

    self.inp = nn.Linear(train_dataset.tensors[0].shape[1], 128)

    # self.bnorm1 = nn.BatchNorm1d(64)
    self.hid1 = nn.Linear(128, 64)

    # self.bnorm2 = nn.BatchNorm1d(32)
    self.hid2 = nn.Linear(64, 32)

    # self.bnorm3 = nn.BatchNorm1d(16)
    self.hid3 = nn.Linear(32, 16)

    self.hid4 = nn.Linear(16, 8)


    self.out = nn.Linear(8,1)


  def forward(self, x):

    x = F.relu(self.inp(x))
    x = F.relu(self.hid1(x))
    x = F.relu(self.hid2(x))
    x = F.relu(self.hid3(x))
    x = F.relu(self.hid4(x))
    x = self.out(x)

    return x
```

```python
model = Model()
```

```python
model
```

```python
learning_rate = 0.01
n_epochs = 300
```

```python
loss_func = nn.BCEWithLogitsLoss()
optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
from tqdm import tqdm
```

```python
losses = []
train_accs = []
test_accs = []

for epoch in tqdm(range(n_epochs)):


  for (X_batch, y_batch) in train_loader:

    batches_losses = []
    batches_train_accs = []

    model.train()

    y_hat = model(X_batch)
    loss = loss_func(y_hat, y_batch)


    optim.zero_grad()
    loss.backward()
    optim.step()

    #store results
    batches_losses.append(loss.detach())
    batches_train_accs.append(100*torch.mean(((y_hat>0) == y_batch).float()).item())

  #after each epoch
  losses.append(np.mean(batches_losses)) #store mean loss value across the different batches for this epoch
  train_accs.append(np.mean(batches_train_accs))


  X_test, y_test = next(iter(test_loader))

  model.eval()
  with torch.no_grad():
      y_pred = model(X_test)

      test_accs.append(100*torch.mean(((y_pred>0) == y_test).float()).item())
```

```python
plt.plot(losses)
```

```python
fig, axs = plt.subplots(1,2, figsize=(17,6))

axs[0].set_title("Losses")
axs[1].set_title("Accuracies")

axs[0].plot(losses,"-rx")
axs[1].plot(train_accs, "-c>", label="Train")
axs[1].plot(test_accs, "-ko" ,label="Test")

axs[1].legend()

fig.show()
```
