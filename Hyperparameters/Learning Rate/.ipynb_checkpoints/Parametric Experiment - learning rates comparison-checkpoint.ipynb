## Parametric Experiments: Learning Rates Comparison

#### Para esto ser un experimento, sobre el mismo set de datos que generemos, crearemos un modelo de DL de clasificación binaria, con la siguiente arquitectura, que será la misma para todos los experimentos, lo que variará será el valor del `learning_rate`

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from IPython import display
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
plt.style.use("dark_background")
```

### - Generamos los datos para la clasificación binaria que se usará en el experimento

```python
n_per_clust = 100 #número de datos que habrán alrededor de cada centro de categoría
blur = 1 #std, indica cómo de "dispersos" estarán los datos

#coordenadas de la nube de datos de cada categoría (solo eje x)
A = [  1, 1 ] #el centro de la categoría A (x=1,y=1)
B = [  5, 1 ] #el centro de la categoría B (x=5,y=1)

# generar datos aleatorios o "ruido" alrededor de cada centro (A,B)
a = [ A[0]+np.random.randn(n_per_clust)*blur , A[1]+np.random.randn(n_per_clust)*blur ]
b = [ B[0]+np.random.randn(n_per_clust)*blur , B[1]+np.random.randn(n_per_clust)*blur ]

# true labels
#para las etiquetas, asignamos ceros a la categoría A, y 1 a la categoría B
labels_np = np.vstack((np.zeros((n_per_clust,1)),np.ones((n_per_clust,1))))

# concatenamos los datos aleatorios en una matriz traspuesta
data_np = np.hstack((a,b)).T

# convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# show the data
fig = plt.figure(figsize=(10,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'bs')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'yo')
plt.title('The qwerties!')
plt.xlabel('qwerty dimension 1')
plt.ylabel('qwerty dimension 2')
plt.show()
```

### - Función para crear el modelo

#### *Como vimos en la imagen de la portada, el modelo toma las 2 características de entrada*

```python
def build_model(learning_rate):
    
    #arquitectura
    model = nn.Sequential(
    
        nn.Linear(2,1), #entrada=2, output=1
        nn.ReLU(),
        nn.Linear(1,1) #output
        #la función sigmoide la incorpora la función de pérdidas en PyTorch llamada BCEWithLogitsLoss (binary crossentropy)
    )
    
    loss_func = nn.BCEWithLogitsLoss()
    
    optimizer = torch.optim.SGD(params=model.parameters(), lr=learning_rate)
    
    return model, loss_func, optimizer
```

### - Función que entrena cada modelo

```python
from tqdm import tqdm
```

#### Nota: a la hora de comparar el output del modelo con las etiquetas reales, creamos el vector de booleanos en base a si las predicciones son > 0, en lugar de ser a partir de > 0.5 (ya que ese es el límite de la función sigmoide). **Sin embargo, al usar la función de pérdida `BCEWithLogitsLoss()`, no definimos explícitamente la función sigmoide al final del modelo, ya que lo hace internamente el framework. Por tanto, la salida del modelo son tan solo valores lineales, que no están mapeados por ninguna función-no-lineal**

```python
n_epochs = 1000

def train(model, progress_bar=True):
    
    losses = torch.zeros(n_epochs) #para almacenar las pérdidas para cada epoch
    
    iterator = tqdm(range(n_epochs)) if progress_bar else range(n_epochs)
    
    for epoch in iterator:
    
        #forward pass
        y_hat = model(data)

        #compute error
        loss = loss_func(y_hat, labels)
        losses[epoch] = loss

        #backward pass
        optim.zero_grad() #clear previous gradients cache
        loss.backward() #compute gradients based on the error
        optim.step() #update parameters
        
    #"final forward pass" - post entrenamiento
    preds = model(data)
    
    #categorizamos las predicciones de salida del modelo a 0s y 1s
    #(validamos las predicciones que sean mayor a 0, y las comparamos con las etiquetas reales creando un vector de booleanos)
    #resultando en ceros y unos dependiendo si el modelo acertó o no, aplicamos la media a esto, y lo convertimos en porcentaje
    total_acc = 100*torch.mean(((preds>0) == labels).float()) #no deberíamos buscar preds > 0.5? --> el output del modelo son valores lineales (en este caso)
    
    return losses, preds, total_acc
```

### - Antes de poner en marcha el experimento paramétrico, testeamos el código una vez

```python
#creamos los objetos necesarios
model, loss_func, optim = build_model(learning_rate=0.1)

#ejecutar el modelo
losses, preds, total_acc = train(model)

#visualizar la precisión
print(f"Final Accuracy: {total_acc} %")

#visualizar la pérdida durante las epochs
plt.plot(losses.detach(),'o',markerfacecolor='w',linewidth=.1)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

### - Hora del experimento!

- #### Creamos un vector de 40 valores linealmente espaciados desde 0.001 a 0.1, que serán los distintos valores de `learning_rates` que ejecutaremos durante el experimento

```python
n_epochs = 1000
learning_rates = np.linspace(0.001, 0.1, 40)

#creamos las matrices que almacenarán los resultados de cada experimento
experiment_accs = [] #precisiones
experiment_losses = np.zeros((len(learning_rates), n_epochs))

#ejecutar el experimento
for idx, lr in enumerate(learning_rates):
        
    #crear y ejecutar modelo
    model, loss_func, optim = build_model(learning_rate=lr)
    losses, preds, total_acc = train(model)
    
    #almacenar resultados
    experiment_accs.append(total_acc)
    experiment_losses[idx,:] = losses.detach()
```

### - Graficar los resultados del experimento

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
fig.suptitle("Parametric Experiment: Learning Rate Comparison",
                fontweight="bold", fontsize=10)

axs[0].plot(learning_rates, experiment_accs, "s-")
axs[0].set_xlabel("Learning Rate")
axs[0].set_ylabel("Accuracy")
axs[0].set_title("Accuracy By Learning Rate")

axs[1].plot(experiment_losses.T)
axs[1].set_title("Losses By Learning Rate")
axs[1].set_xlabel("Epoch No.")
axs[1].set_ylabel("Loss")

plt.show();
```

- ### En este caso, vemos cómo (1a gráfica) las precisiones van "variando" a lo largo de los distintos valores de `learning_rates`. Apreciamos desde precisiones del 100% en todo el rango de valores que hemos probado de learning_rate, y también precisiones "no tan deseadas" o e incluso del 50% o menos, es decir, el modelo no consigue "aprender" por lo que acierta puramente a suerte. Puede que en dichos casos, el modelo hubiese quedado atrapado en mínimos locales con una pérdida considerablemente alta

- ### Por otro lado, en la (2da gráfica) vemos cómo los valores de `learning_rate` que sí obtuvieron un buen valor, fueron decreciendo con el paso de las epochs, y otros valores de learning_rate (con la precisión más baja) cuyas líneas fueron prácticamente "rectas" o constantes durante el paso de las epochs, indicando que el modelo no "aprendía" nada, y por tanto, la pérdida apenas cambiaba

- ### Además, hay casos donde al final, el modelo acabó consiguiendo acabar en una buena solución a pesar de haber comenzado con una pérdida constante, como es el caso en la segunda gráfica, de líneas donde a partir de 200 epochs por ejemplo, comenzaron a bajar sus pérdidas y acercarse asintóticamente a 0

## Conclusión; viendo que su precisión "o es alta", o "no es alta", excepto en algunos casos que obtiene algo "intermedio", podemos ver cómo depende de la "suerte" que tenga la inincialización de los pesos, por lo que la **suerte es un gran factor**

### - Otra forma de evaluar el desempeño de los modelos:

```python
#proporción de ejecuciones donde el modelo tuvo al menos un '70%' de precisión <-- ese valor podemos establecerlo en el que deseemos
#usando torch.tensor() pasamos de array NumPy a Tensor de PyTorch (dividimos el booleano de precisiones > que ese valor) y dividimos entre total valores
sum(torch.tensor(experiment_accs) > 70) / len(experiment_accs)
```

## - **Nuevo Experimento** ("Meta Experiment" para obtener resultados más confiables)-

### Ejecutaremos el experimento anterior 50 veces y promediaremos los resultados
### - Debido a la gran variabilidad que obtenemos en los resultados dependiendo de la suerte, haciendo que no podamos observar la verdadera "relación" entre la precisión y el learning_rate

*Una ejecución del meta-experimento corresponde a ejecutar el experimento anterior N veces*

```python
#(recuerda que cada uno toma distintas inincializaciones aleatorias de pesos)
num_exps = 50 #nº de veces a iterar a través del experimento

metaexp_accs = np.zeros((num_exps, len(learning_rates)))

#para tomar menos tiempo de cómputo, haremos 500 epochs
n_epochs = 500

#ejecutamos el experimento anterior, N veces
for exp in range(num_exps):
    
    #dentro de cada experimento, ejecutamos un modelo distinto para cada combinación de learning rate
    for idx, lr in enumerate(learning_rates):
        
        #crear y ejecutar modelo
        model, loss_func, optim = build_model(learning_rate=lr)
        losses, preds, total_acc = train(model, progress_bar=False)

        #almacenar resultados
        metaexp_accs[exp, idx] = total_acc #guardamos la precisión final
    
    if exp%5 == 0 and exp!=0:
        print(f"Meta-experiment {(exp/num_exps)*100}% sucessfully completed")
    elif exp==num_exps-1:
        print(f"Meta-experiment 100% sucessfully completed")
    #una vez ejecute el modelo para cada valor de learning_rate especificado, completará "un experimento"
```

```python
#graficar resultados
#para cada distinto valor de learning_rate, la media de las precisiones de 50 ejecuciones con dicho valor
plt.plot(learning_rates, np.mean(metaexp_accs, axis=0), "s-")
plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.title("Accuracy By Learning Rate (mean of 50 runs for each lr value)")
plt.show()
```

### Ahora tenemos una vista más clara, y confiable sobre cómo se comportan la precisión en base los distintos valores de `learning_rates`, ya que para cada uno hemos ejecutado 50 experimentos con dicho valor, y obtenido la media

- ### Vemos que, generalmente valores muy pequeños de learning_rate no conducen a un "buen aprendizaje" (pues se queda "atascado" sin dar apenas pasos), mientras que valores más altos de learning_rate facilitan el "aprendizaje", obteniendo mayores medias de precisión

## **Quizás, al estar las precisiones en el rango de alrededor 50% a 100%, la media no era la mejor medida a ver el desempeño del modelo en base a multiples ejecuciones**
