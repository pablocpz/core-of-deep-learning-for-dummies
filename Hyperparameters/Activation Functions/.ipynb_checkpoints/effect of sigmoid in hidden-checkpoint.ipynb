```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
X = np.linspace(-5, 5, 500).reshape(-1,1)
y = np.sin(X)
```

```python
X = torch.from_numpy(X).float()
y = torch.from_numpy(y).float()
```

```python
plt.plot(X, y);
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
train_data = TensorDataset(X_train, y_train)
test_data = TensorDataset(X_test, y_test)
```

```python
batch_size = 16 #suelen ser potencias de 2

train_loader = DataLoader(train_data,
                         batch_size=batch_size,
                         shuffle=True)

test_loader = DataLoader(test_data,
                        batch_size=test_data.tensors[0].shape[0])
#train_data.tensors[0] accede a las características X
#para predicciones del modelo, los lotes son individuales
```

```python
class Model(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.inp = nn.Linear(X_train.shape[1], 128)
        self.hid1 = nn.Linear(128, 32)
        self.hid2 = nn.Linear(32, 16)
        self.hid3 = nn.Linear(16, 8)
        self.hid4 = nn.Linear(8, 4)
        self.out = nn.Linear(4, 1)
        
    def forward(self, x):
        x = F.sigmoid(self.inp(x))
        x = F.sigmoid(self.hid1(x))
        x = F.sigmoid(self.hid2(x))
        x = F.sigmoid(self.hid3(x))
        x = F.sigmoid(self.hid4(x))
        x = self.out(x)
        return x
```

```python
model = Model()
```

```python
learning_rate = 0.01

optim = torch.optim.SGD(params=model.parameters(), lr=learning_rate)
loss_func = nn.MSELoss()
```

```python
n_epochs = 100
```

```python
from tqdm import tqdm
```

```python
model.parameters()
```

```python
predictions, out_gradients = torch.zeros((n_epochs, 500, 1)), torch.zeros((n_epochs, 1))
losses = []

for epoch in tqdm(range(n_epochs)):
    
    batch_losses = []
    
    for (X_batch, y_batch) in train_loader:
        y_hat = model(X_batch)
        loss = loss_func(y_hat, y_batch)
        
        optim.zero_grad()
        loss.backward()
        optim.step()
        
        batch_losses.append(loss.detach())
    
    losses.append(np.mean(batch_losses))
    out_gradients[epoch] = model.out.weight.grad.mean() #store current mean gradient of last layer
    
    preds = model(X)
    predictions[epoch] = preds    
```

```python
plt.plot(losses);
```
