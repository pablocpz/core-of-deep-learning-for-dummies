## Cross Validation in regression example

Dado que no experimentaremos con distintas arquitecturas de modelos (por ejemplo variando el learning_rate, batch size...etc) sino que tan solo creamos uno solo, usaremos `train set` y `test set`

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

## Creando el dataset

```python
N = 100 #tendrá 100 observaciones
X = torch.rand(N,1)
y = X + torch.rand(N, 1)

plt.plot(X, y, "s");
```

## Creando el modelo

```python
class Model(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.inp = nn.Linear(1,1)
        self.out = nn.Linear(1,1)
        
    def forward(self, x):
        
        x = F.relu(self.inp(x))
        x = self.out(x)
        
        return x
```

```python
model = Model()
```

```python
loss_func = nn.MSELoss() 

learning_rate = 0.01

optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

## Dividiendo los datos

Lo haremos manualmente como demostración, elegiremos aleatoriamente 80 valores que se encuentren en el rango del tamaño de observaciones (100 filas)

```python
train_idxs = np.random.choice(range(N), 80, replace=False)
train_bool = np.zeros(N, dtype=bool) #inincializamos un vector de True(s)/False(s)

#estableceremos en True los índices aleatorios obtenidos antes, (serán las filas pertencecientes a X_train)
train_bool[train_idxs] = True

print(X[train_bool].shape)
print(X[~train_bool].shape) #los que no son True, pertenecerán a testing
```

```python
X_train, y_train = X[train_bool], y[train_bool]
X_test, y_test = X[~train_bool], y[~train_bool]
```

## Entrenando el modelo

```python
n_epochs = 500

for epoch in range(n_epochs):
    
    #forward pass
    y_hat = model(X_train)

    #compute error
    loss = loss_func(y_hat, y_train)

    #backward pass
    optim.zero_grad() #clear previous gradients cache
    loss.backward() #compute gradients based on the error
    optim.step() #update parameters
    
```

## Calculando las pérdidas post-entrenamiento

```python
y_test_pred = model(X_test)
test_loss = (y_test_pred-y_test).pow(2).mean() #MSE manualmente

print(f"Final training loss: {loss.detach():.2f}")
print(f"Final test loss: {test_loss.detach():.2f}")
```

## Graficando las predicciones

```python
#graficaremos ambos sets de datos (train y test) y sus predicciones
plt.figure(figsize=(10,6))
y_train_pred = model(X_train).detach().numpy()

plt.plot(X, y, "rs", label="Entire Data Set")
# plt.plot(X_train, y_train, "rs");
plt.plot(X_train, y_train_pred, "bs", markerfacecolor="aqua", label="Train Predictions");
plt.plot(X_test, y_test_pred.detach(), "ro", markerfacecolor="white", label="Test Predictions")

plt.legend();
plt.show();
# plt.plot(X_test, pr)
```

```python
N = 100 #tendrá 100 observaciones
X = torch.rand(N,1)
y = X + torch.rand(N, 1)

plt.plot(X, y, "s");
```

```python
class Model(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.inp = nn.Linear(1,1)
        self.out = nn.Linear(1,1)
        
    def forward(self, x):
        
        x = F.relu(self.inp(x))
        x = self.out(x)
        
        return x
```

```python
model = Model()
```

```python
loss_func = nn.MSELoss() 

learning_rate = 0.01

optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
train_idxs = np.random.choice(range(N), 80, replace=False)
train_bool = np.zeros(N, dtype=bool) #inincializamos un vector de True(s)/False(s)

#estableceremos en True los índices aleatorios obtenidos antes, (serán las filas pertencecientes a X_train)
train_bool[train_idxs] = True

print(X[train_bool].shape)
print(X[~train_bool].shape) #los que no son True, pertenecerán a testing
```

```python
X_train, y_train = X[train_bool], y[train_bool]
X_test, y_test = X[~train_bool], y[~train_bool]
```

```python
n_epochs = 500
train_losses, test_losses = [], []

for epoch in range(n_epochs):
    
    #forward pass
    y_hat = model(X_train)

    #compute error
    loss = loss_func(y_hat, y_train)

    #backward pass
    optim.zero_grad() #clear previous gradients cache
    loss.backward() #compute gradients based on the error
    optim.step() #update parameters
    
    train_loss = (model(X_train)-y_train).pow(2).mean()
    test_loss = (model(X_test)-y_test).pow(2).mean() #MSE manualmente
    
    train_losses.append(train_loss.detach()), test_losses.append(test_loss.detach())
```

```python
y_test_pred = model(X_test)
test_loss = (y_test_pred-y_test).pow(2).mean() #MSE manualmente

print(f"Final training loss: {loss.detach():.2f}")
print(f"Final test loss: {test_loss.detach():.2f}")
```

```python
plt.figure(figsize=(10,6));
plt.plot(range(0,500), train_losses, color="gold", label="Train Loss")
plt.plot(range(0,500), test_losses, color="aqua", label="Test Loss");
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend();
```

Vemos, como es de esperar, en este caso, que la pérdida en el entrenamiento es ligeramente inferior al de testing, (es decir, desempeña levemente mejor), y también podemos apreciar como ambas pérdidas van decreciendo conjuntamente hasta aproximarse o estacionarse asintóticamente en un valor de pérdida en torno a las 200 epochs
