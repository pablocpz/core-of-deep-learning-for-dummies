```python
def linear(x):

    return np.random.rand(1,1) + np.dot(x,
                                       np.random.rand(x.shape[1], 1)) 

def nonlinear(x):

    return np.maximum(0,x)
```

```python
X = np.linspace(-10,10).reshape(-1,1)

import matplotlib.pyplot as plt

l1, l2, l3 = linear(X), linear(X), linear(X)

l1 = np.random.rand(1,1)*4*X -20

l2 = np.random.rand(1,1)*1.25*X + 3

l3 = np.random.rand(1,1)*-25*X -8
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,6))

axs[0].plot(X, l1)
axs[0].set_title("lin1 = mx+b")
axs[1].plot(X, l2)
axs[2].plot(X, l3);
```

```python
h1, h2, h3 = nonlinear(l1), nonlinear(l2), nonlinear(l3)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,6))

axs[0].plot(X, h1)
axs[0].set_title("h1 = act(lin1)")
axs[1].plot(X, h2)
axs[2].plot(X, h3);
```

```python
o1 = np.random.rand(1,1)*-1*h1 -10
o2 = np.random.rand(1,1)*5*h2 +10
o3 = np.random.rand(1,1)*1.25*h3 +np.random.rand(1,1)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,6))

axs[0].plot(X, o1)
axs[0].set_title(r"$o_1 = \phi \cdot h_1$")
axs[1].plot(X, o2)
axs[2].plot(X, o3);
```
